##### 2.9. создание нового поля вычисленного на основе значений других полей:

2.9.1 через выражение на базе имеющихся колонок

```python
# добавим столбец, в котором будут значение = длина / диаметр
df1['Rings/RingsMax'] = df1['Rings']/df1['Rings'].max()
df1
```

2.9.2 через DataFrame.apply

```python
# добавим столбец "условный объем" = pi*r^2*h

df1['Volume'] = df1.apply(lambda row: row['Length']*np.pi*((row['Diameter']/2)**2), axis = 1)
df1
```

2.9.3 через Series.apply

```python
df1['Round_Vol'] = df1['Volume'].apply(lambda val: round(val, 2))
df1
```
